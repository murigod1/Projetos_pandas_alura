```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df = pd.read_csv(url)
df.head(20)
```

```python
df.fillna(0)
```

```python
df['Pontos_extras'] = df['Notas']*0.40
df.head()
```

```python
df['Notas_finais'] = df['Notas'] + df['Pontos_extras']
df.head()
```

```python
df['Aprovado_final'] = df['Notas_finais'].apply(lambda x: 'Aprovado' if x >= 6 else 'Rejeitado')
df.head(20)
```

```python
df['Notas_finais'] = df['Notas_finais'].apply(lambda x: 10 if x >= 10 else x)
df.head(20)
```

```python
df['Aprovado'] = df['Aprovado'].apply(lambda x: 'Aprovado' if x == True else 'Rejeitado')
df.head(20)
```

```python
df['Salvos'] = df['Aprovado'] != df['Aprovado_final']
df.head(20)
```
