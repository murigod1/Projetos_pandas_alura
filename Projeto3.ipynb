```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
display(df)
```

```python
selecao = (df['Nome'] == 'Alice') | (df['Nome'] == 'Carlos')
leave_class = df[selecao].index
```

```python
df.drop(leave_class, axis=0, inplace=True)
```

```python
df = df.loc[(df['Aprovado'] == True)]
```

```python
df.sort_values(by='Notas',ascending=False)
```

```python
df = df.replace({'Notas':7}, {'Notas': 8})
df.sort_values(by='Notas',ascending=False)
```

```python
df.to_csv('alunos_aprovados.csv', index=None)
```
